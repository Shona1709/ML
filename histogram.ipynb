```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming 'data.csv' is the name of your dataset file
# Adjust the filename and column names according to your dataset
data = pd.read_csv('/data_covid.csv')

# Grouping the data by country and summing the total tests done for each country
country_total_tests = data.groupby('country')['tests_done'].sum()

# Plotting the histogram
plt.figure(figsize=(12, 6))
country_total_tests.plot(kind='bar')
plt.title('Total Tests Done by Country')
plt.xlabel('Country')
plt.ylabel('Total Tests')
plt.xticks(rotation=90)  # Rotate x-axis labels for better visibility
plt.tight_layout()
plt.show()
```
